Cruce CX
* cruce de 2 padres y mantiene la posición de ciertos valores que se heredan de forma ciclica

```python
import random

def cruce_cx(p1,p2, target_col=0):
    # Aseguramos que ambos padres tengan el mismo tamaño
    size = len(p1)
    
    # inicialización del hijo con valores None
    offspring = [None] * size

    #punto de inicio aleatorio
    start = random.randint(0, size - 1)
    
    #guarda  el ciclo de los índices que serán intercambiados
    indices = []
    
    # Iniciamos el ciclo con el valor del gen en el punto de inicio
    current_index = start
    while current_index not in indices:
        indices.append(current_index)
        # Encuentra la posición en el otro padre del valor actual del primer padre
        current_index = p2.index(p1[current_index])
    
    # rellena  los valores del ciclo desde el primer padre
    for index in indices:
        offspring[index] = p1[index]

    # completa  el resto con los valores del segundo padre
    for i in range(size):
        if offspring[i] is None:
            offspring[i] = p2[i]
    
    return offspring

# Ejemplo de uso
p1 = [1, 2, 3, 4, 5, 6, 7, 8]
p2 = [4, 1, 2, 8, 7, 6, 3, 5]

hijo = cruce_cx(p1, p2)
print("Padre 1:", p1)
print("Padre 2:", p2)
print("Hijo:   ", hijo)
```

```python
import numpy as np
import random

def cruce_cx2(parents_slice):
    p1 = parents_slice[0]
    p2 = parents_slice[1]

    size = len(p1)

    # Initialize offspring with None values
    offspring = np.full(size, None)

    # Random start index
    start = random.randint(0, size - 1)

    # List to store cycle indices
    indices = []

    current_index = start
    while current_index not in indices:
        indices.append(current_index)
        # Position in p2 of the current value in p1
        current_index = np.where(p2 == p1[current_index])[0][0]

    # Fill offspring with values from p1 (cycle)
    for index in indices:
        offspring[index] = p1[index]

    # Complete the rest with values from p2
    for i in range(size):
        if offspring[i] is None:
            offspring[i] = p2[i]

    return offspring
```

Mutación Scramble

```python
def mutacion_scramble(individuo):
    n_cromos = len(individuo)
    
    # selección de dos indices aleatorios
    idx1,idx2 = np.sort(np.random.choice(range(n_cromos),size=2,replace=False))
    # segmento a mutar 
    scramble_part = individuo[idx1:idx2+1]
    print(f'subrista:{scramble_part}')

    # desorden del segmente
    np.random.shuffle(scramble_part)
    print(f'subrista shuffle:{scramble_part}')
    
    # selección de un índice aleatorio para insertar la parte mutada 
    
    #reemplazo de la parte del individuo con el segmento desordenado
    individuo[idx1:idx2+1] = scramble_part

    return individuo 
```

```python
# Ejemplo: Si tienes una población de 100 individuos, cada uno con 18 genes
population_size = 100
gene_length = 18

# Generamos una población inicial aleatoria de 100 individuos con permutaciones de 18 genes
population = [np.random.permutation(gene_length) for _ in range(population_size)]

# mutación por scramble a cada individuo en la población
mutation_rate = 0.1  # Probabilidad de mutación

for i in range(population_size):
    if np.random.rand() < mutation_rate:  # Aplica la mutación con una cierta probabilidad
        print(f"antes de la mutación {i}: {population[i]}")
        population[i] = mutacion_scramble(population[i])
        print(f"después de la mutación {i}: {population[i]}")
```

```python
# esta función ya realiza la separación adecuada para las tercias sin importar si 
# es par o no la longitud de los cromosoams 
def cruce_dos_corte(padres):
    n_cromos = len(padres[0, 2])
    p1 = padres[0, 2]
    p2 = padres[1, 2]

    # Convertimos las cadenas de cromosomas en listas de genes
    p1 = list(map(str, p1))  # Convertir a lista de caracteres
    p2 = list(map(str, p2))

    # calcula la longitud de cada parte
    part_length = n_cromos // 3
    remaining = n_cromos % 3

    # divide los cromosomas en tres partes 
    parts_p = []
    parts_m = []
    start = 0
    for i in range(3):
        end = start + part_length
        if i < remaining:
            end += 1
        parts_p.append(p1[start:end])
        parts_m.append(p2[start:end])
        start = end

    h1 = f"{parts_m[0]}{parts_p[1]}{parts_m[2]}"
    h2 = f"{parts_p[0]}{parts_m[1]}{parts_p[2]}"

    return h1, h2
```

```python
# sólo para cuando son de dos digitos los genes 
def cruce_dos_corte3(padres, gene_length=2):
    n_cromos = len(padres[0, 2])
    p1 = padres[0, 2]
    p2 = padres[1, 2]

    # Convertimos las cadenas de cromosomas en listas de genes
    p1 = [p1[i:i+gene_length] for i in range(0, n_cromos, gene_length)]
    p2 = [p2[i:i+gene_length] for i in range(0, n_cromos, gene_length)]

    # calcula la longitud de cada parte
    part_length = len(p1) // 3
    remaining = len(p1) % 3

    # divide los cromosomas en tres partes 
    parts_p = []
    parts_m = []
    start = 0
    for i in range(3):
        end = start + part_length
        if i < remaining:
            end += 1
        parts_p.append(p1[start:end])
        parts_m.append(p2[start:end])
        start = end

    h1 = f"{parts_m[0]}{parts_p[1]}{parts_m[2]}"
    h2 = f"{parts_p[0]}{parts_m[1]}{parts_p[2]}"

    return h1, h2
```
